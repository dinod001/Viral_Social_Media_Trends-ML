### 01. Import dependencies


```python
import umap
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

### 02. Load dataset


```python
df = pd.read_csv('../../../EDA/data/processed/Standarlized_data.csv')
df.head()
```

```python
X_umap = df.copy()
```

### 03. Apply UMAP


```python
# UMAP for 2D visualization
umap_2d = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1, verbose=True)
umap_scores_2d = umap_2d.fit_transform(X_umap)
print(f"UMAP 2D embedding shape: {umap_scores_2d.shape}")
```

### 04. Visualize UMAP 2D projection


```python
# 2D UMAP visualization
plt.figure(figsize=(10, 7))
plt.scatter(umap_scores_2d[:, 0], umap_scores_2d[:, 1], alpha=0.6, s=50)
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.title('UMAP 2D Projection')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 05. Apply UMAP for higher dimensions (for clustering)


```python
# UMAP for 3D visualization
umap_3d = umap.UMAP(n_components=3, random_state=42, n_neighbors=15, min_dist=0.1, verbose=True)
umap_scores_3d = umap_3d.fit_transform(X_umap)
print(f"UMAP 3D embedding shape: {umap_scores_3d.shape}")
```

### 06. Visualize UMAP 3D projection


```python
# Optional: 3D visualization
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(umap_scores_3d[:, 0], umap_scores_3d[:, 1], umap_scores_3d[:, 2], s=50, alpha=0.6)
ax.set_xlabel('UMAP Component 1')
ax.set_ylabel('UMAP Component 2')
ax.set_zlabel('UMAP Component 3')
ax.set_title('UMAP 3D Projection')
plt.tight_layout()
plt.show()
```

### 08. Saving Dataset


```python
n_components_to_save = 3

umap_top = pd.DataFrame(
    umap_scores_3d[:, :n_components_to_save],
    columns=[f'PC{i+1}' for i in range(n_components_to_save)]
)

output_path = '../data/UMAP_components.csv'
umap_top.to_csv(output_path, index=False)
print("Saved UMAP components successfully!")
```
